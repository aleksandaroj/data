```python
ds = pd.read_csv("countries.csv")
```

```python
dse = pd.read_csv("country_codes.csv")
```

```python
pd.merge(ds, dse, on="country")
```

```python
joined = ds.join(dse)
```

```python
merged = pd.merge(ds, dse, on="country", how="left")
```

```python
merged
```

```python
afg = merged[merged.code == "AFG"]
afg
```

```python
srb = merged[merged.code == "SRB"]
srb
```

```python
plt.plot(afg.year, afg.population / 10 ** 6)
plt.plot(srb.year, srb.population / 10 ** 6)
plt.xlabel = "Godina"
plt.ylabel = "populacija (u milionima)"
plt.legend = (["Afganistan", "Srbija"])
plt.show()
```
