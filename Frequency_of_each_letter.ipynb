```python
import matplotlib.pyplot as plt
```

```python
alphabet = 'abcdefghijklmnopqrstuvwxyzäöüßđčćž'
len(alphabet)
```

```python
text_eng = '''
The procession to St George's Chapel for the funeral service will be led by the band of the Grenadier Guards.

The band will be followed by the Major General's party and military chiefs of staff, including the Chief of the Defence Staff General Sir Nicholas Carter, and First Sea Lord Admiral Tony Radakin.

Admiral Radakin paid tribute to the duke's naval career, saying "He was one of us."

Also in the procession behind the Land Rover will be Brigadier Archie Miller Bakewell, a personal protection officer, two pages and two valets.

Brigadier Bakewell was Prince Philip's private secretary and right-hand man for more than a decade, having started working for the duke back in 2010.

The Dean of Windsor, David Conner, will conduct the funeral service and give the commendation as the coffin is lowered into the royal vault.
'''
```

```python
text_ger = '''
Grund dafür ist in erster Linie das fehlende Fachpersonal. So haben seit Beginn der Pandemie rund 9.000 Pflegekräfte ihren Beruf verlassen. Ein Verlust, der auch durch die zunehmende Immunisierung von Pflegekräften sowie von Ärztinnen und Ärzten im Vergleich zur zweiten Welle nicht aufgefangen werden kann.

Außerdem wurde zu Jahresbeginn der Personalschlüssel für Pflegekräfte auf Intensivstationen von 2,5 Betten pro Pflegekraft auf zwei Betten gesenkt. "Die Maßnahme ist an sich gut, da es den Pflegekräften mehr Zeit gibt, sich um Patienten zu kümmern", erklärt DIVI-Pressesprecher Albrecht. "Die Kehrseite ist jedoch: Betten, für die es kein Personal gibt, müssen eben 'gesperrt' werden."

Die Folge: Es wird zunehmend enger auf den Intensivstationen. Von den zurzeit verfügbaren 23.917 Intensivbetten sind 21.053 belegt, der Anteil der noch freien Betten liegt bei zwölf Prozent. Zum Vergleich: Auf dem Höhepunkt der zweiten Welle (Stichtag 16.12.2020) lag der Anteil der freien Betten bei knapp 17 Prozent.
'''
```

```python
text_srb = '''
Srbija bi početkom maja trebalo da dobije prve isporuke vakcina preko programa donacija koji finansira EU, saznaje Tanjug u Briselu.

Povodom koordinacije isporuka vakcina iz EU za zemlje Zapadnog Balkana, održan je i video-sastanak evropskog komesara za proširenje Olivera Varheljija i ministara zemlja Zapadnog Balkna.

"Sa ministrima spoljnih poslova i ministrima za zaduženim za evropske integracije razgovarao sam o poslednjim koracima koji treba da osiguraju isporuke vakcina protiv kovida 19 regionu kako bi se pomogla vakcinacija medicinskih radnika i najugroženijih grupa", objavio je Vrahelji na Tviteru.

Iz EU naglašavaju da nastavljaju svoju podršku partnerima na Zapadnom Balkanu.

Vakcine iz EU, čija bi prva isporuka trebalo da bude početkom maja, biće finasiranje u potunosti iz fondova EU, za šta je izdvojeno 70 miliona evra bespovratih sredstva.

Istovremeno, EU je jedan od najvećih donatora i međunarodnog instrumenta za nabavku vakcina Kovaks, preko koga je u Srbiju već stiglo 56.700 doza vakcine Astara-Zeneke, navodi se u Briselu.
'''
```

```python
letter_counts_eng = [text_eng.count(l) for l in alphabet]

letter_counts_ger = [text_ger.count(l) for l in alphabet]

letter_counts_srb = [text_srb.count(l) for l in alphabet]
```

```python
letter_colors_eng = plt.cm.hsv([0.8*i/max(letter_counts_eng) for i in letter_counts_eng])

letter_colors_ger = plt.cm.hsv([0.8*i/max(letter_counts_ger) for i in letter_counts_ger])

letter_colors_srb = plt.cm.hsv([0.8*i/max(letter_counts_srb) for i in letter_counts_srb])
```

```python
plt.bar(range(34), letter_counts_eng, color=letter_colors_eng)
plt.xticks(range(34), alphabet)
plt.tick_params(axis='x', bottom=False)
plt.title('Frequency of each letter (eng)')
plt.show()
```

```python
plt.bar(range(34), letter_counts_ger, color=letter_colors_ger)
plt.xticks(range(34), alphabet)
plt.tick_params(axis='x', bottom=False)
plt.title('Frequency of each letter (ger)')
plt.show()
```

```python
plt.bar(range(34), letter_counts_srb, color=letter_colors_srb)
plt.xticks(range(34), alphabet)
plt.tick_params(axis='x', bottom=False)
plt.title('Frequency of each letter (srb)')
plt.show()
```
